```python
import os
from dotenv import load_dotenv
load_dotenv(override=True, dotenv_path="../.env.local")
my_api_key = os.getenv("OPENAI_API_KEY")
my_api_key[:5], my_api_key[-10:]
```

```python
from openai import OpenAI
client = OpenAI(api_key=my_api_key)
```

```python
# prompt = "Who was Napoleon Bonaparte and why was he important in European history?"
prompt = "If humans could live for 200 years, how would society change?"
# prompt = "Write a short story about a city that floats in the sky."
# prompt = "Why do people fear artificial intelligence, and are those fears justified?"
# prompt = "What would happen if all electricity stopped working for one week?"

response = client.chat.completions.create(
    model="gpt-5-nano", #"gpt-5-turbo", #"gpt-5", #"gpt-5-mini", #"gpt-5-nano"
    messages=[
        {"role": "system", "content": "You are a helpful assistant. Answer the user's questions concisely."},
        {"role": "user", "content": prompt}
        # {"role": "user", "content": f"Here is the context: \n\ {context} \n\. \n\n Now, answer the question: {prompt}")
    ]
)
print (response)
print (response.choices[0].message.content)
```
