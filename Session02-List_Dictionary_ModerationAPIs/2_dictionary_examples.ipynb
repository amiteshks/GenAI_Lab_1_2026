```python
person1 = {
    "id": 1,
	"name": "Alice",
	"age": 5,
	"is_student": False
}

person2 = {
    "id": 2,
	"name": "Tim",
	"age": 30,
	"is_student": False
}


person3 = {
    "id": 3,
	"name": "Mark",
	"age": 35,
	"is_student": False
}

people_in_myhouse = [person1, person2, person3]
average_age = 0
total_age = 0
for person in people_in_myhouse:
    age = person["age"]
    print(f"Person {person['name']} is {age} years old.")
    total_age += age

average_age = total_age / len(people_in_myhouse)
average_age
    
```
